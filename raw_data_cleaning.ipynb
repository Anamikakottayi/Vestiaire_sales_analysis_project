```python
import pandas as pd

df = pd.read_csv("vestiaire.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python

df['product_category'] = df['product_category'].fillna('unknown')
df['product_material'] = df['product_material'].fillna('unknown')
df['product_color'] = df['product_color'].fillna('unknown')
df['seller_username'] = df['seller_username'].fillna('unknown')
```

```python
df = df.drop(columns=['product_description', 'product_keywords'], errors='ignore')
```

```python
bool_cols = ['sold', 'reserved', 'available', 'in_stock', 'should_be_gone']

for col in bool_cols:
    df[col] = df[col].astype(int)
```

```python
numeric_cols = [
    'price_usd', 'seller_price', 'seller_earning',
    'buyers_fees', 'seller_products_sold',
    'seller_num_products_listed', 'seller_community_rank',
    'seller_num_followers', 'seller_pass_rate'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Fill missing numeric values
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
text_cols = ['brand_name', 'product_category', 'product_color', 'seller_country']

for col in text_cols:
    df[col] = df[col].str.lower().str.strip()
```

```python
df = df[(df['price_usd'] > 0) & (df['price_usd'] < 10000)]
```

```python
df = df.drop_duplicates()
```

```python
# Price range for dashboard
df['price_range'] = pd.cut(
    df['price_usd'],
    bins=[0, 100, 500, 2000, 10000],
    labels=['very_low', 'low', 'medium', 'high']
)
```

```python
# Seller conversion rate (powerful feature)
df['conversion_rate'] = df['seller_products_sold'] / df['seller_num_products_listed']
```

```python
print(df.info())
print(df.isnull().sum())
```

```python

df['has_cross_border_fees'] = df['has_cross_border_fees'].astype(str).str.lower().str.strip()


df['has_cross_border_fees'] = df['has_cross_border_fees'].replace({
    'yes': 1,
    'no': 0,
    'true': 1,
    'false': 0,
    'nan': 0,    
    'none': 0,
    '': 0
})

df['has_cross_border_fees'] = pd.to_numeric(df['has_cross_border_fees'], errors='coerce')


df['has_cross_border_fees'] = df['has_cross_border_fees'].fillna(0)


df['has_cross_border_fees'] = df['has_cross_border_fees'].astype(int)
```

```python
df['has_cross_border_fees'].value_counts()
```

```python
df['shipping_days'] = df['usually_ships_within'].str.extract('(\d+)')
df['shipping_days'] = pd.to_numeric(df['shipping_days'], errors='coerce')

df['shipping_days'] = df['shipping_days'].fillna(df['shipping_days'].median())
```

```python
df['product_season'] = df['product_season'].fillna('unknown')
```

```python
df['conversion_rate'] = df['conversion_rate'].fillna(0)
```

```python
df = df.drop(columns=['usually_ships_within'])
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df.to_csv("cleaned_vestiaire.csv", index=False)
```
